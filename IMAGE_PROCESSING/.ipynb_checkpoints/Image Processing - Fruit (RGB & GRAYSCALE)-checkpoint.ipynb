# Data Preprocessing

## Create a grayscale folders for grayscale images conversion from the original dataset

```python
def load_and_convert_to_grayscale(dataset_path, class_name):
    class_path = os.path.join(dataset_path, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    for image_file in image_files:
        image_path = os.path.join(class_path, image_file)
        original_image = cv2.imread(image_path)
        grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
        yield original_image, grayscale_image, class_name, image_file

def save_grayscale_images(output_path, original_image, grayscale_image, class_name, image_name):
    class_output_path = os.path.join(output_path, class_name)
    os.makedirs(class_output_path, exist_ok=True)
    output_image_path = os.path.join(class_output_path, image_name)
    cv2.imwrite(output_image_path, grayscale_image)

def process_dataset(dataset_path, output_dataset_path):
    for class_name in os.listdir(dataset_path):
        if os.path.isdir(os.path.join(dataset_path, class_name)):
            for original, grayscale, class_name, image_name in load_and_convert_to_grayscale(dataset_path, class_name):
                save_grayscale_images(output_dataset_path, original, grayscale, class_name, image_name)

# Provide the path to your training dataset
train_dataset_path = '/Users/HP/Documents/STUDIES/PYTHONCODES/DATASETS/FRUITS&VEGGIE/train/Fruits'
output_train_dataset_path = '/Users/HP/Documents/STUDIES/PYTHONCODES/DATASETS/FRUITS&VEGGIE(GRAYSCALE)/train/Fruits'

# Process the training dataset
process_dataset(train_dataset_path, output_train_dataset_path)

# Provide the path to your validation dataset
val_dataset_path = '/Users/HP/Documents/STUDIES/PYTHONCODES/DATASETS/FRUITS&VEGGIE/validation/Fruits'
output_val_dataset_path = '/Users/HP/Documents/STUDIES/PYTHONCODES/DATASETS/FRUITS&VEGGIE(GRAYSCALE)/validation/Fruits'

# Process the validation dataset
process_dataset(val_dataset_path, output_val_dataset_path)
```

```python
# Load grayscale datasets and display the images
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the grayscale dataset
gray_training_set = tf.keras.utils.image_dataset_from_directory(
    '/Users/HP/Documents/STUDIES/PYTHONCODES/DATASETS/FRUITS&VEGGIE(GRAYSCALE)/train/Fruits',
    labels='inferred',
    label_mode='categorical',
    class_names=None,
    color_mode='grayscale',
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False
)

gray_validation_set = tf.keras.utils.image_dataset_from_directory(
    '/Users/HP/Documents/STUDIES/PYTHONCODES/DATASETS/FRUITS&VEGGIE(GRAYSCALE)/validation/Fruits',
    labels='inferred',
    label_mode='categorical',
    class_names=None,
    color_mode='grayscale',
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False
)

# Function to display images from the grayscale dataset
def show_images(dataset, num_images=5):
    plt.figure(figsize=(15, 15), dpi=100)
    for images, labels in dataset.take(1):
        for i in range(num_images):
            ax = plt.subplot(1, num_images, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"), cmap='gray')
            plt.title(f'Class: {tf.argmax(labels[i])}')
            plt.axis("off")

# Display images from the grayscale training set
show_images(gray_training_set, num_images=8)

plt.show()
```

# Grayscale CNN Model

```python
# This mode use grayscale dataset as input
cnn_gray = tf.keras.models.Sequential()
cnn_gray.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,padding='same',activation='relu',input_shape=[64,64,1]))
cnn_gray.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'))
cnn_gray.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'))
cnn_gray.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn_gray.add(tf.keras.layers.Dropout(0.25))

cnn_gray.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding='same',activation='relu'))
cnn_gray.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'))
cnn_gray.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'))
cnn_gray.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn_gray.add(tf.keras.layers.Dropout(0.25))

cnn_gray.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,padding='same',activation='relu'))
cnn_gray.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
cnn_gray.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn_gray.add(tf.keras.layers.Dropout(0.25))

cnn_gray.add(tf.keras.layers.Flatten())
     
cnn_gray.add(tf.keras.layers.Dense(units=512,activation='relu'))
cnn_gray.add(tf.keras.layers.Dense(units=256,activation='relu'))   
cnn_gray.add(tf.keras.layers.Dropout(0.25)) #To avoid overfitting
     

#Output Layer
cnn_gray.add(tf.keras.layers.Dense(units=11,activation='softmax'))
```

```python
cnn_gray.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
cnn_gray.summary()
```

```python
training_history_fruit_gray = cnn_gray.fit(x=gray_training_set,validation_data=gray_validation_set,epochs=60)
```

```python
print("Train set Accuracy: {} %".format(training_history_fruit_gray.history['accuracy'][-1]*100))
print("Validation set Accuracy: {} %".format(training_history_fruit_gray.history['val_accuracy'][-1]*100))
```

#  RGB dataset in CNN Model

```python
# Now we use RGB dataset in here
cnn2 = tf.keras.models.Sequential()
cnn2.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,padding='same',activation='relu',input_shape=[64,64,3]))
cnn2.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'))
cnn2.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'))
cnn2.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn2.add(tf.keras.layers.Dropout(0.25))

cnn2.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding='same',activation='relu'))
cnn2.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'))
cnn2.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'))
cnn2.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn2.add(tf.keras.layers.Dropout(0.25))

cnn2.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,padding='same',activation='relu'))
cnn2.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
cnn2.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn2.add(tf.keras.layers.Dropout(0.25))

cnn2.add(tf.keras.layers.Flatten())
     
cnn2.add(tf.keras.layers.Dense(units=512,activation='relu'))
cnn2.add(tf.keras.layers.Dense(units=256,activation='relu'))   
cnn2.add(tf.keras.layers.Dropout(0.25)) #To avoid overfitting
     

#Output Layer
cnn2.add(tf.keras.layers.Dense(units=11,activation='softmax'))
```

```python
cnn2.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
cnn2.summary()
```

```python
training_history_fruit_rgb = cnn2.fit(x=training_set_rgb,validation_data=validation_set_rgb,epochs=40)
```

```python
print("Train set Accuracy: {} %".format(training_history_fruit_rgb.history['accuracy'][-1]*100))
print("Validation set Accuracy: {} %".format(training_history_fruit_rgb.history['val_accuracy'][-1]*100))
```
